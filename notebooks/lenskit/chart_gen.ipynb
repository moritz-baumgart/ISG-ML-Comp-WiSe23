```python
import joblib
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
```

```python
data = {
    '85/15 split with\ndata after 2016': [0.9656301465557162, 1.23006],
    '85/15 split with\ndata after 2018': [0.9555878694303016, 1.2247],
}

df = pd.DataFrame.from_dict(data)

df = df.melt(var_name='Config', value_name='RMSE')

df['type'] = ['TTSplit Validation', 'Kaggle'] * (len(df) // 2)

df
```

```python
g = sns.barplot(df, x='RMSE', y='Config', hue='type', palette=sns.color_palette('Set2'))
for i in g.containers:
    g.bar_label(i, label_type='center', fontsize=8)
plt.yticks(rotation=90, va='center')
#plt.xscale('log')
plt.legend(loc=8, bbox_to_anchor=(0.5, -0.3))

plt.show()
```
